```python
#Importar bibliotecas que se ocupan
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#cargar csv
df = pd.read_csv('P1-RegresionLineal.csv')
```

```python
#mostrar grafico
plt.figure(figsize=(10, 6))
plt.scatter(df['equipos'], df['costo'])
plt.title('Datos Originales')
plt.xlabel('Equipos Afectados')
plt.ylabel('Costo del Incidente')
plt.grid(True)
plt.show()
```

```python
#convertir datos  a valores reales (1000,10000)
df['equipos'] = df['equipos'] * 1000
df['costo'] = df['costo'] * 1000
```

```python
#grafico con valores reales 
plt.figure(figsize=(10,6))
plt.scatter(df['equipos'],df['costo'],color='purple')
plt.title('datps reales')
plt.xlabel('equipos afectados(reales)')
plt.ylabel('costo del incidente(reales)')
plt.grid(True)
plt.show()
```

```python
#Generar el modelo de regresion lineal
lin_reg=LinearRegression()
lin_reg.fit(df['equipos'].values.reshape(-1,1),df['costo'].values)
```

```python
#Mostrar theta
theta_1=lin_reg.coef_[0]
print(f"Theta 1(pendiente):{theta_1}")
```

```python
x_min_max = np.array([[df['equipos'].min()], [df['equipos'].max()]])
y_pred = lin_reg.predict(x_min_max)
```

```python
# Graficar el modelo con el conjunto de datos
plt.plot(df['equipos'], df['costo'], "b.")
plt.plot(x_min_max, y_pred, "g-")
plt.xlabel("Equipos")
plt.ylabel("Coste del incidente")
plt.title("Regresin lineal con datos")
plt.show()
```

```python
#  Generar un archivo CSV con 20 entradas de datos (valores simulados)
nuevo_df = pd.DataFrame({
    'equipos': np.linspace(df['equipos'].min(), df['equipos'].max(), 20),
    'costo': lin_reg.predict(np.linspace(df['equipos'].min(), df['equipos'].max(), 20).reshape(-1, 1))
})

# Guardar el nuevo archivo CSV
nuevo_df.to_csv("nuevo_datos.csv", index=False)
print("Nuevo archivo CSV con 20 entradas generado.")
```

```python
# Leer el archivo generado y pasarlo a un DataFrame
nuevo_df_leido = pd.read_csv("nuevo_datos.csv")
print("Datos del nuevo archivo CSV:")
print(nuevo_df_leido.head())
```

```python
# Graficar la prediccin de los 20 valores
plt.plot(nuevo_df_leido['equipos'], nuevo_df_leido['costo'], "r-", label="Prediccin")
plt.scatter(nuevo_df_leido['equipos'], nuevo_df_leido['costo'], color='pink', label="Datos CSV")
plt.xlabel("Equipos")
plt.ylabel("Coste")
plt.title("Prediccin con 20 valores")
plt.legend()
plt.show()
```
